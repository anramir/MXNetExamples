```python
import mxnet as mx
from mxnet import nd
```

```python
probabilities = nd.ones(6) / 6
sample = nd.sample_multinomial(probabilities)
```

```python
print(sample)
```

```python
nd.sample_multinomial(probabilities, shape=(4, 5))
```

```python
rolls = nd.sample_multinomial(probabilities, shape=(1000))
```

```python
counts = nd.zeros((6, 1000))
total = nd.zeros(6)
for i, roll in enumerate(rolls):
    total[int(roll.asscalar())] += 1
    counts[:, i] = total
```

```python
total / 1000
```

```python
x = nd.arange(1000).reshape((1, 1000)) + 1
estimates = counts / x
print(estimates[:, 100])
```

```python
%matplotlib inline

from matplotlib import pyplot as plt
plt.plot(estimates[0, :].asnumpy(), label="Estimated P(die=0)")
plt.plot(estimates[1, :].asnumpy(), label="Estimated P(die=1)")
plt.plot(estimates[2, :].asnumpy(), label="Estimated P(die=2)")
plt.plot(estimates[3, :].asnumpy(), label="Estimated P(die=3)")
plt.plot(estimates[4, :].asnumpy(), label="Estimated P(die=4)")
plt.plot(estimates[5, :].asnumpy(), label="Estimated P(die=5)")
plt.axhline(y = 0.16666, color = "black", linestyle = "dashed")
plt.legend()
plt.show()
```
